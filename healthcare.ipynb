ML MODELS 

1. KNN MODEL

```python
x=data1.drop("Outcome",axis=1)
```

```python
y=data1["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
print(X_train.shape)
print(X_test.shape) 
print(y_train.shape) 
print(y_test.shape) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train,y_train)
```

```python
y_pred_test=knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_test)
```

```python
y_pred_train=knn_model.predict(X_train)
accuracy_score(y_train,y_pred_train)
```

```python
from  sklearn.metrics import classification_report
from sklearn.metrics import  confusion_matrix
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
print(confusion_matrix(y_test,y_pred_test))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
probs=knn_model.predict_proba(X_test)
probs = probs[:, 1]
auc_lr=roc_auc_score(y_test,probs)
print('AUC: %.3f' % auc_lr)
fpr, tpr, thresholds = roc_curve(y_test, probs) 
plt.plot([0, 1], [0, 1], linestyle='--') 
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC (Receiver Operating Characteristics) Curve")
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import average_precision_score
precision, recall, thresolds = precision_recall_curve(y_test,probs)
y_test_pred=knn_model.predict(X_test)
F1_score=f1_score(y_test,y_test_pred)
auc_lr=auc(recall,precision)
ap=average_precision_score(y_test,probs)
print("f1_score=%.3f auc_lr=%.3f ap=%.3f" %(F1_score,auc_lr,ap))
plt.plot([0 , 1], [0.5,0.5],linestyle= '--')
plt.plot(recall, precision, marker='.')
plt.xlabel("Recall")
plt.ylabel("precision");
```
